# Project Lifecycle

Every Mad House project moves through six stages. The stage determines what the project can receive: free experimentation in lab, human review in prod, and so on.

The `incubate.sh` tool manages this lifecycle. This notebook explains the model and how to use it.

## The Six Stages

| Stage | What it means | What agents can do |
|-------|--------------|-------------------|
| **concept** | Idea exists, no code | Discuss, spec, plan |
| **prototype** | Proof of concept, disposable | Commit freely, break things |
| **building** | Real build, scope locked | Commit freely, respect scope |
| **shipped** | Live in production | Require review for changes |
| **maintained** | Stable, receiving only bug fixes | Require review for anything non-trivial |
| **archived** | Dead. Read-only. | Read only |

A project can be promoted forward (concept > prototype > building > shipped) or demoted backward (shipped > building for major rework, shipped > archived for decommission).

## Using `incubate.sh`

Located at `~/dev/mad-house/tooling/workspace-tools/incubate.sh`.

```bash
# Start a new project
incubate.sh new my-idea

# List all projects and their stages
incubate.sh list

# Advance a project to the next stage
incubate.sh promote my-idea

# Advance to a specific stage
incubate.sh stage my-idea building

# Mark as shipped
incubate.sh ship my-idea

# Archive
incubate.sh archive my-idea

# Get info on a project
incubate.sh info my-idea
```

## Stage Gates

Each promotion requires a deliberate decision. The questions to answer before advancing:

**concept > prototype**
- Is the idea specific enough to build a spike?
- Is there a concrete thing to test?

**prototype > building**
- Did the prototype prove the core assumption?
- Is scope locked? (No scope-locked = not ready)
- Is there a repo, a README, and a Dockerfile?

**building > shipped**
- Does it run in Docker without manual steps?
- Are all secrets in env vars, not code?
- Has it run for 24h in staging without crashes?
- Is the pre-launch checklist complete?

**shipped > maintained**
- Has it been stable in production for at least one week?
- Is there a runbook for common issues?

**any stage > archived**
- Is it definitively dead? (Not "on pause" or "maybe later")
- Archive is irreversible in intent. If there is any chance of revival, it stays where it is.

## Lab vs. Prod

The lab/prod split is a hard boundary, not a convention.

| | Lab (`~/dev/mad-house/lab/`) | Prod (top-level, Coolify) |
|---|---|---|
| Stage | concept, prototype | shipped, maintained |
| Commits | Free, agents can commit | Human review required |
| Deploy | Never to production | Coolify, human-triggered |
| Experiments | Yes | No |
| Breaking changes | Fine | Require rollback plan |

A project that is building can live in either location depending on whether it needs Coolify access. When it ships, it moves to a top-level repo and a Coolify project.

## When to Archive

Archive when:
- The project has not been touched in 3+ months and there is no plan to touch it
- The need it solved no longer exists
- A replacement was shipped
- It was a prototype that proved the idea wrong

Do not archive because it is "paused" or "low priority". Archive is a commitment to done.

**On GitHub:** archive the repo. This makes it read-only and removes it from the active repo list.

**Locally:** remove from `~/dev/mad-house/`. The code is in GitHub; you do not need a local copy of dead projects.

## Project Manifest

The workspace manifest tracks all active projects. Located at `~/dev/mad-house/tooling/agents.manifest.yaml`.

```yaml
projects:
  - name: chopsticks
    stage: maintained
    repo: madebymadhouse/chopsticks
    deployed: true
    description: Live Discord bot. Community infrastructure.

  - name: snek
    stage: building
    repo: madebymadhouse/snek
    deployed: false
    description: Discord bot for dev tooling.
```

Keep this up to date. It is what `workspace-audit.sh` reads to determine what should be running.
