# Org and Repo Setup

How to set up a GitHub organization, create repos that follow Mad House standards,
and keep the org clean over time.

## Org Structure

Three orgs, three partitions. They never blend.

| Org | Purpose | What goes here |
|-----|---------|----------------|
| `madebymadhouse` | Build house products | Bots, agents, tools, experiments |
| `orinadus-systems` | Orinadus platform | Urchin, orinadus-platform, brand assets |
| `samhcharles` | Personal | Dotfiles, personal-dashboard, profile |

If you are not sure which org a project belongs to, answer: who is the customer?
- Mad House community or internal tooling: `madebymadhouse`
- Orinadus platform users: `orinadus-systems`
- Only you: `samhcharles`

## Creating a New Repo

```bash
# Via GitHub CLI
gh repo create madebymadhouse/my-project \
  --description "One sentence description" \
  --public \
  --clone

cd my-project
```

### Visibility rules

- Tooling, bots, experiments: **public** (build in the open)
- Internal context files, secrets, personal notes: **private**
- Platform infra (Orinadus): **private** unless it is a public product

Default to public. If you are not sure whether to make it public, ask: would a stranger finding this repo be a problem?

## Standard Files Every Repo Needs

```
README.md             What it is, how to install, how to use, how to contribute
LICENSE               MIT for open source, omit for private
.gitignore            Standard for the language + always include: .env, AGENTS.md, CLAUDE.md
CHANGELOG.md          Updated at every release
.env.example          All env vars with placeholder values (commit this, not .env)
Dockerfile            If it runs as a service
```

Optional but good:
```
CONTRIBUTING.md       How to contribute (good for public repos)
.github/
  workflows/          CI actions
  ISSUE_TEMPLATE/     Bug and feature request templates
```

## README Template

```markdown
# Project Name

One sentence description.

## What it does

2-3 sentences. What problem it solves and for whom.

## Quick start

```bash
git clone https://github.com/madebymadhouse/project-name
cd project-name
cp .env.example .env
# Fill in .env
docker compose up
```

These are also in `~/.gitignore_global`, but per-repo gitignore is the defense-in-depth layer.

## Branch Protection

For any repo with a production deployment:

1. GitHub > Settings > Branches > Add rule
2. Branch name pattern: `main`
3. Enable:
   - Require pull request reviews before merging (1 reviewer minimum)
   - Require status checks to pass (if CI exists)
   - Do not allow force pushes

For solo projects, at minimum: disable force push on `main`.

## Keeping the Org Clean

Monthly org audit:

```bash
# List all repos with archive status
gh repo list madebymadhouse --json name,isArchived,updatedAt \
  --jq '.[] | "\(.isArchived | if . then "ARCHIVED" else "LIVE   " end) \(.updatedAt[0:10]) \(.name)"' \
  | sort -k3
```

For each live repo not touched in 3+ months: decide to archive or note an active reason it exists.

**Archive rule:** if a project is done and there is no plan to touch it, archive it on GitHub. It becomes read-only, removed from the active list, but the code is preserved.
